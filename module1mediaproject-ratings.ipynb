```python
def jprint(obj):
    # create a formatted string of the Python JSON object
    text = json.dumps(obj, sort_keys=True, indent=4)
    print(text)
```

```python
movie_ratings = []
for i in clean_data['imdbid']:
    url = 'http://www.omdbapi.com/?i={}&apikey=18d87115'.format(i)
    response = requests.get(url)
    movie_ratings.append(jprint(response.json()))
```
